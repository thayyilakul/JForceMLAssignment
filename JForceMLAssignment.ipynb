```python
tf.random.set_seed(42)

model_1_embedding = layers.Embedding(input_dim = max_vocab_length,
                                    output_dim = 128,
                                    embeddings_initializer = 'uniform',
                                    input_length = max_length,
                                    name = 'embedding_1')

inputs = layers.Input(shape = (1, ), dtype = 'string')
x = text_vectorizer(inputs)
x = model_1_embedding(x)
x = layers.LSTM(64)(x)
x = layers.Dense(4, activation = 'relu')(x)
x = layers.Dense(4, activation = 'relu')(x)
outputs = layers.Dense(len(product_count), activation = 'softmax')(x)
model_1 = tf.keras.Model(inputs, outputs, name = 'model_1_LSTM')
```

```python
tf.random.set_seed(42)

model_2_embedding = layers.Embedding(input_dim = max_vocab_length,
                                    output_dim = 128,
                                    embeddings_initializer = 'uniform',
                                    input_length = max_length,
                                    name = 'embedding_2')

inputs = layers.Input(shape = (1, ), dtype = 'string')
x = text_vectorizer(inputs)
x = model_2_embedding(x)
x = layers.LSTM(64)(x)
x = layers.Dense(10, activation = 'relu')(x)
x = layers.Dense(10, activation = 'relu')(x)
x = layers.Dense(10, activation = 'relu')(x)
x = layers.Dense(10, activation = 'relu')(x)
outputs = layers.Dense(len(sub_product_count), activation = 'softmax')(x)
model_2 = tf.keras.Model(inputs, outputs, name = 'model_2_GRU')
```

```python
model_2.compile(loss = 'sparse_categorical_crossentropy',
                optimizer = 'adam',
                metrics = ['accuracy'])
```

```python
model_2.summary()
```

```python
model_2.fit(train2_sentences,
            train_sub_product_labels,
            epochs = 5,
            validation_data = (val2_sentences, val_sub_product_labels))
```

```python
# For predicting product
test_samples = random.sample(sorted(train2_sentences), 10)
for test_sample in test_samples:
    prob = model_2.predict(np.array([test_sample], dtype = 'object'))
    arg_pred = prob.argmax()
    print(f'Prediction : {sub_product_count[arg_pred]}')
    print(f'Text: \n{test_sample}\n')
```

```python
def argmax(model, sentence):
  prod_prob = model.predict(np.array([sentence], dtype = 'object'))
  return prod_prob.argmax()

def response(product_arg_pred, sub_prod_arg_pred, sentence):
    print(f'Product Prediction: {product_count[product_arg_pred]}')
    print(f'Sub-Product Prediction : {sub_product_count[sub_prod_arg_pred]}')
    print(f'Text: \n{sentence}\n')
```

```python
def predict_sentences(sentence = False):
  if(sentence):
    product_arg_pred = argmax(model_1, sentence)
    sub_prod_arg_pred = argmax(model_2, sentence)
    response(product_arg_pred, sub_prod_arg_pred, sentence)
  else:
    test_samples = random.sample(sorted(train2_sentences), 10)
    for test_sample in test_samples:
      product_arg_pred = argmax(model_1, test_sample)
      sub_prod_arg_pred = argmax(model_2, test_sample)
      response(product_arg_pred, sub_prod_arg_pred, test_sample)
```
